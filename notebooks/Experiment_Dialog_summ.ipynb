```python
import json
INPUT_FILE = "pc_usr_data.json"
with open(INPUT_FILE, 'r') as f:
    data = json.load(f)
```

```python
import random
SEED=42
random.seed(SEED)
random.shuffle(data)
```

```python
from transformers import pipeline
summarizer = pipeline('summarization', model="google/pegasus-cnn_dailymail")
```

```python
for idx in range(20):
    if len(data[idx]['context'].split(' ')) > 100:
        print("-"*10)
        # print(data[idx]['context'])
        
        utterances = list(filter(lambda x: len(x.strip()) > 0, data[idx]['context'].split('\n')))
        new_utterances = []
        for _, instance in enumerate(utterances):
            new_utterances.append(f"A{_%2 + 1}: {instance}")

        old_history, curr_utterance = "<n>".join(new_utterances[:-1]), new_utterances[-1]
        
        print("History")
        print(f"Length: {len(old_history.split(' '))} words")
        print(old_history.replace("<n>", "\n"))
        print("")
        
        summary = summarizer(old_history, max_length=len(old_history.split(' ')) // 2, min_length=30, do_sample=False)
        summary_text = summary[0]['summary_text']
        # summary_text = summary_text.replace("<n>", "\n")

        print(f"Summary")
        print(summary_text)
        print("")

        print("Current Utterance")
        print(curr_utterance)
        print("")

        print("Response without summary:")
        print("") 
        
        print("Response with summary:")
        print("")

        # split_data = list(filter(lambda x: len(x.strip()) > 0, data[idx]['context'].split('\n')))
        # for _, instance in enumerate(split_data):
        #     print(f"A{_%2+1}: {instance}") 
        # print("")
        # print(data[idx]['fact'])

        for response in data[idx]['responses']:
            if response['model'] in ['Original Ground Truth', 'New Human Generated']:
                print(response['model'])
                print(response['response'])
```
